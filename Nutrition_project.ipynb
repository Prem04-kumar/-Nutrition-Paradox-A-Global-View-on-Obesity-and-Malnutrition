```python
#Countries where female obesity exceeds male by large margin (same year)


q9="""SELECT
    country,
    year,
    AVG(CASE WHEN gender = 'Female' THEN mean_estimate END) AS female_obesity,
    AVG(CASE WHEN gender = 'Male' THEN mean_estimate END) AS male_obesity,
    (
      AVG(CASE WHEN gender = 'Female' THEN mean_estimate END) -
      AVG(CASE WHEN gender = 'Male' THEN mean_estimate END)
    ) AS obesity_gap
FROM obesity_table
GROUP BY country, year
HAVING
    AVG(CASE WHEN gender = 'Female' THEN mean_estimate END) IS NOT NULL
AND AVG(CASE WHEN gender = 'Male' THEN mean_estimate END) IS NOT NULL
AND (
      AVG(CASE WHEN gender = 'Female' THEN mean_estimate END) -
      AVG(CASE WHEN gender = 'Male' THEN mean_estimate END)
    ) > 5
ORDER BY obesity_gap DESC
LIMIT 10;"""

df_obesity_q9=pd.read_sql(q9,engine)
df_obesity_q9

```

```python
# 1. Obesity vs malnutrition comparison by country(any 5 countries)
join_q1="""SELECT
    t1.country,
    t1.year,
    AVG(t1.mean_estimate) AS malnutrition_rate,
    AVG(t2.mean_estimate) AS obesity_rate
FROM malnutrition_table t1
INNER JOIN obesity_table t2
    ON t1.country = t2.country
    AND t1.year = t2.year
WHERE t1.country IN (
    'India',
    'Nigeria',
    'Brazil',
    'United States',
    'China'
)
GROUP BY t1.country, t1.year
ORDER BY t1.country, t1.year;""" 

df_join_q1=pd.read_sql(join_q1,engine)
df_join_q1

df_melted = df_join_q1.melt(
    id_vars=["country", "year"],
    value_vars=["malnutrition_rate", "obesity_rate"],
    var_name="health_metric",
    value_name="rate"
)

fig = px.line(
    df_melted,
    x="year",
    y="rate",
    color="health_metric",
    facet_col="country",
    facet_col_wrap=2,
    markers=True,
    title="Malnutrition vs Obesity Trends Across Countries"
)

fig.update_layout(
    template="plotly_white",
    height=800
)

fig.show()
```

```python
# 2.Gender-based disparity in both obesity and malnutrition
join_q2="""SELECT
    t1.gender,
    AVG(t1.mean_estimate) AS avg_malnutrition_rate,
    AVG(t2.mean_estimate) AS avg_obesity_rate
FROM malnutrition_table t1
INNER JOIN obesity_table t2
    ON t1.country = t2.country
    AND t1.year = t2.year
    AND t1.gender = t2.gender
GROUP BY t1.gender
ORDER BY t1.gender;"""

df_join_q2=pd.read_sql(join_q2,engine)
df_join_q2


df_long = df_join_q2.melt(
    id_vars="gender",
    value_vars=["avg_malnutrition_rate", "avg_obesity_rate"],
    var_name="health_condition",
    value_name="rate"
)

plt.figure(figsize=(8, 6))

sns.barplot(
    data=df_long,
    x="gender",
    y="rate",
    hue="health_condition"
)

plt.title("Gender-based Disparity in Obesity and Malnutrition")
plt.xlabel("Gender")
plt.ylabel("Average Rate")
plt.legend(title="Health Condition")
plt.tight_layout()
plt.show()
```

```python
# 3.Region-wise avg estimates side-by-side(Africa and America)
join_q3="""SELECT
    t1.region,
    AVG(t1.mean_estimate) AS avg_malnutrition_rate,
    AVG(t2.mean_estimate) AS avg_obesity_rate
FROM malnutrition_table t1
INNER JOIN obesity_table t2
    ON t1.country = t2.country
    AND t1.year = t2.year
WHERE t1.region IN ('Africa', 'Americas')
GROUP BY t1.region
ORDER BY t1.region;""" 
df_join_q3=pd.read_sql(join_q3,engine)
df_join_q3
```

```python
# 4.Countries with obesity up & malnutrition down

join_q4="""SELECT
    T1.Country,
    (MAX(T1.Mean_Estimate) - MIN(T1.Mean_Estimate)) AS obesity_change,
    (MAX(T2.mean_estimate) - MIN(T2.mean_estimate)) AS malnutrition_change
FROM obesity_table T1
JOIN malnutrition_table T2
    ON T1.Country = T2.Country
GROUP BY T1.Country
HAVING obesity_change > 0 AND malnutrition_change < 0;""" 

df_join_q4= pd.read_sql(join_q4,engine)
df_join_q4
```

```python
# 5.Age-wise trend analysis
obesity_age = (
    df_obesity
    .groupby("age_group")["Mean_Estimate"]
    .mean()
    .reset_index()
)
obesity_age["Type"] = "Obesity"

malnutrition_age = (
    df_malnutrition
    .groupby("age_group")["Mean_Estimate"]
    .mean()
    .reset_index()
)
malnutrition_age["Type"] = "Malnutrition"

combined_age = pd.concat([obesity_age, malnutrition_age])
```
